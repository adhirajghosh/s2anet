```python
from mmdet.apis import init_detector, inference_detector, show_result_pyplot
import mmcv
```

```python
config_file = '../configs/dota/s2anet_r50_fpn_1x_dota.py'
# download the checkpoint from model zoo and put it in `checkpoints/`
checkpoint_file = '../work_dirs/s2anet_r50_fpn_1x_dota/latest.pth'
```

```python
# build the model from a config file and a checkpoint file
model = init_detector(config_file, checkpoint_file, device='cuda:0')
```

```python
# test a single image
img = '../data/dota_1024/trainval_split/images/P0420__1.0__0___94.png'
result = inference_detector(model, img)
```

```python
# show the results
show_result_pyplot(img, result, model.CLASSES)
```
